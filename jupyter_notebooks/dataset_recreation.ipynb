```python
import pandas as pd
import csv

# assign meaningful column names if you like
col_names = [
    'tweet_id', 'created_at', 'user_id', 'coords',
    'retweet_id', 'text',
    'score1', 'score2', 'score3', 'score4'
]

df = pd.read_csv(
    'twitteraae_all_aa',
    sep='\t',
    header=None,
    names=col_names,
    engine='python',               
    quoting=csv.QUOTE_NONE,        # treat all quotes as literal
    on_bad_lines='skip',           # drop any lines that still fail parsing
    dtype=str                      # read everything as string to avoid surprises
)
```

```python
# Simple one-liner filter
filtered_df = df[df['text'].str.split().str.len() > 10]

print(filtered_df.shape)
```

## Removing double back slashes and non-ascci characters

```python
import re

# your validity-checking function
def is_valid_sentence(sentence: str) -> bool:
    # Non-ASCII and control characters
    pattern1 = re.compile(r'[^\x00-\x7F]+|[\x00-\x1F\x7F]')
    # Backslashes
    pattern2 = re.compile(r'[\\]')
    return not (pattern1.search(sentence) or pattern2.search(sentence))

mask = filtered_df['text'].apply(is_valid_sentence)
```

keep only rows where textp passes the check

```python
df_clean = filtered_df[mask].reset_index(drop=True)
```

```python
df_clean.shape
```

```python
from typing import List, Union

def find_text_indices(df: pd.DataFrame, substring: str, case: bool = False) -> List[int]:
    """
    Return a list of row indices where `df['text']` contains `substring`.
    
    Parameters
    ----------
    df : pd.DataFrame
        Your DataFrame, must have a 'text' column.
    substring : str
        The substring to search for.
    case : bool, default False
        If False, the search is case-insensitive.
    
    Returns
    -------
    List[int]
        List of integer indices.
    """
    # Build the boolean mask
    mask = df['text'].str.contains(substring, case=case, na=False)
    # Extract matching indices
    return df.index[mask].tolist()
```
